```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.ticker as mtick
import numpy as np
```

```python
wine_df = pd.read_csv('C:\\Hims\\winequality-white.csv',sep=';')
wine_df.head()
```

```python
ax = sns.distplot(wine_df['alcohol'])#, kde=False, rug=True)
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1))
plt.annotate('42% wines have alcohol content > 9',xy=(9.6,0.42),xytext=(11,0.40), arrowprops=dict(facecolor='r'))
```

```python
with sns.axes_style('ticks'):
    sns.set(color_codes=True)
    sns.kdeplot(wine_df['alcohol'], shade=True, color='r')
    sns.despine(offset=50)
```

```python
sns.set_palette('cubehelix')
sns.jointplot(wine_df['free sulfur dioxide'],wine_df['total sulfur dioxide'], kind='hex',xlim={0,130},ylim={0,300})
```

```python
sns.pairplot(wine_df,height=5, diag_kind='kde', kind='reg', palette='Paired')
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(wine_df.corr(),square=False, annot=True, fmt='0.2f', cmap='summer', vmax=0.8, vmin=-0.8)
```

```python
sns.set_style('whitegrid')
with sns.color_palette('Blues'):
    sns.lmplot(x='residual sugar',y='density', data=wine_df)
```

### Lmplot to show Mean and Std Deviation for discrete values

```python
sns.lmplot(x='quality',y='alcohol', data=wine_df, x_estimator=np.mean)
```

```python
sns.lmplot(x='pH',y='fixed acidity', data=wine_df, row='quality')#col='quality')
```

```python
sns.regplot(x='pH',y='fixed acidity',data=wine_df)
```

### Categorical plots

```python
plt.figure(figsize=(15,10))
sns.stripplot(x='quality',y='alcohol',data=wine_df, jitter=True)
```

```python
plt.figure(figsize=(15,10))
sns.swarmplot(x='quality',y='alcohol',data=wine_df)
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='quality',y='alcohol',data=wine_df)
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(x='quality',y='alcohol',data=wine_df)
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(x='quality',y='alcohol',data=wine_df)
sns.swarmplot(x='quality',y='alcohol',data=wine_df, color='c')
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='alcohol',data=wine_df)
plt.xticks(rotation=90)
```

```python
sns.set_style('dark')
sns.set_context('paper')
plt.figure(figsize=(15,5))
sns.pointplot(x='quality',y='alcohol',data=wine_df)
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=wine_df, orient='h')
```
